```python
def er_graph(N,p):
    '''Generate an ER Graph.'''
    L=nx.Graph()
    L.add_nodes_from(list(range(N)))
    for node1 in L.nodes():
        for node2 in L.nodes():
            if(node1 <node2 and bernoulli.rvs(p=p)): 
                L.add_edge(node1,node2)
    return L
```

```python
nx.draw(er_graph(50,0.08), node_size=40, node_color="gray")
```

```python
L.degree()
```

```python
def plot_degree_distribution(G):
    degree_sequence = [d for n,d in G.degree()]
    plt.hist(degree_sequence, histtype="step")
    plt.xlabel("Degree $k$")
    plt.ylabel("$P(K)$")
    plt.title("Degree Distribution")
```

```python
E=er_graph(50,0.08)
plot_degree_distribution(E)
```

```python
#Above most connected node has 8 connections. Most number of nodes were somehwere around 2 and 3 edges.
```

```python
Y=er_graph(500,0.08)
plot_degree_distribution(Y)
```

```python
#Here much more degrees since much more nodes. The P(k) is there as how many nodes have that degree. No less than around 20 
#degrees and no more than around 60.
```

```python
G1=er_graph(500,0.08)
plot_degree_distribution(G1)
G2=er_graph(500,0.08)
plot_degree_distribution(G2)
G3=er_graph(500,0.08)
plot_degree_distribution(G3)
#This is an example of degrees distributions drawn from three different Erdos-Renyi graphs.
```

```python
#The files to be loaded are adjacency marticies where 1 signified an edge between two edges and 0 means none.
import numpy as np
A1=np.loadtxt("Village1.csv", delimiter=",")
A2=np.loadtxt("Village2.csv", delimiter=",")
```

```python
L1=nx.to_networkx_graph(A1)
L2=nx.to_networkx_graph(A2)
#Need to convert adjacency matrices to graph objects.
```

```python
def basic_net_stats(G):
    print("Number of nodes: %d" % G.number_of_nodes())
    print("Number of edges: %d" % G.number_of_edges())
    number_of_degrees=[d for n,d in G.degree()]
    print("Average degree %.2f" % np.mean(number_of_degrees)) #.2f is 2 decimal points
```

```python
basic_net_stats(L1)
```

```python
basic_net_stats(L2)
```

```python
plot_degree_distribution(L1)
plot_degree_distribution(L2)
#Shows here that relatively large amount have few connections (good relations with others in village) and only few have a lot.
#Compared to ER graphs, these are not at all symmetrical, even with large amounts of nodes. Tail extends quite to the right and
#suggests ER graphs are not good representations of real world networks. They can be used as a reference graph by comparing
#their properties to emperical social networks.
```

```python
#LCC is largest connected componenet
L1_LCC= max((L1.subgraph(c) for c in nx.connected_components(L1)), key=len) #key is len since max len is used since len=no. of 
#connected nodes in subgraph and therefore largest component. 
L2_LCC= max((L2.subgraph(c) for c in nx.connected_components(L2)), key=len)
```
