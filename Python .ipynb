```python
#Shallow copy-> Creates Compound Object 'x' copy and then references to all objects x has references to. 
#So if x objects change it changes.
```

```python
#Deep copy-> Creates Compound Object 'x' copy and then recursively inserts copies into it of original objects. 
#So, creates copies of the objects attached to compound objects as well and refeences those so now obejcts aren't interdependent. 
```

```python
import copy
x = [1,[2]]
y = copy.copy(x)
z = copy.deepcopy(x)
y is z
```

```python
#^^ Explanation: copy.copy creates a shallow copy, which constructs a new object whose contents refer to the original object 
#to the greatest extent possible. In contrast, copy.deepcopy creates a deep copy, creating both a new variable name as well as 
#a new object with new content, but equal values to the original. Therefore, y and z refer to different objects.
```

```python
import math
def distance(x,y):
    s=(y[0]-x[0])**2 + (y[1]-x[1])**2
    return math.sqrt(s)
```

```python
def in_circle(x, origin = [0,0]):
    if distance(x,origin)<1:
        return True
    else:
        return False
```

```python
def mavg(x, n_neighbors=1):
    mavg=[]
    i=0
    while i<len(x):
        sumv=[x[i]]
        j = n_neighbors
        while j>0:
            if  (i-j)<0:
                sumv.append(x[i])
                sumv.append(x[i+j])
                j-=1
            elif (i+j)>len(x):
                sumv.append(x[i])
                sumv.append(x[i-j])
                j-=1
            else:
                sumv.append(x[i-j])
                sumv.append(x[i+j])
                j-=1
        mavg_i=sum(sumv)/len(sumv)
        mavg.append(mavg_i)
        i+=1
    return mavg
```

```python
delta_X=np.random.normal(0,1,(2,5))
delta_X
```

```python
plt.plot(delta_X[0], delta_X[1], "go")
```

```python
EE= np.cumsum(delta_X, axis=1) 
EE
#axis=1 since adding the columns of array to get cumilative for x axis and y axis in 2 arrays. This is done since each step is 
#an addition of the previous displacements. THIS IS RANDOM WALKS.
```

```python
plt.plot(EE[0], EE[1], "ro-");
```

```python
X_0=np.array([[0],[0]]) #To concatonate (0,0) into EE array from random walk to start at origin.
```

```python
EE=np.concatenate((X_0,np.cumsum(delta_X, axis=1)), axis=1) 
EE
#tuples of each of the np arrays to concatenate, and axis=1 to add the two 0's column wise.
```

```python
plt.plot(EE[0], EE[1], "ro-");
```

```python
delta_X=np.random.normal(0,1,(2,100)) #2 rows and 100 columns. 100 variables to add later. The larger no., the more walks.
X_0=np.array([[0],[0]])
EE=np.concatenate((X_0,np.cumsum(delta_X, axis=1)), axis=1) 
plt.plot(EE[0], EE[1], "bs-");
```
